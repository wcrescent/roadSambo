```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
df = pd.read_csv('data/105a1a2.CSV', encoding = 'big5')
df
```

```python
df_male = df[df['性別']==1]
df_female = df[df['性別']==2]

print('男性事故件數:', len(df_male))
print('男性所占比例:', len(df_male)/51692*100, '%')
print('女性事故件數:', len(df_female))
print('女性所占比例:', len(df_female)/51692*100, '%')
```

```python
age_male = []
count_male = []
age = []
for i in range(0,104): 
    age.append(i)
    count_male.append(len(df_male[ df_male['年齡']==i]))

age_dict = { '交通事故總計': count_male
}
male_age = pd.DataFrame(age_dict)
```

```python
plt.figure(figsize=(16, 8))
plt.xlim(0, 103)
plt.ylim(0, 1400)
plt.xlabel('age')
plt.ylabel('count')
l1, = plt.plot(male_age)
plt.legend(handles=[l1], labels=['male'],  loc='upper right')
plt.show()
```

```python
count_female = []
for i in range(0,104): 
    count_female.append(len(df_female[df_female['年齡']==i]))

age_dict = {'交通事故總計': count_female
}
female_age = pd.DataFrame(age_dict)

```

```python
plt.figure(figsize=(16, 8))
plt.xlim(0, 103)
plt.ylim(0, 1400)
plt.xlabel('age')
plt.ylabel('count')
l2, = plt.plot(female_age ,color='red')
plt.legend(handles=[l2], labels=['female'],  loc='upper right')
plt.show()
```

```python
plt.figure(figsize=(16, 8))
plt.xlim(0, 103)
plt.ylim(0, 1400)
plt.xlabel('age')
plt.ylabel('count')
l1, =plt.plot(male_age)
l2, =plt.plot(female_age ,color='red')
plt.legend(handles=[l1, l2], labels=['male', 'female'],  loc='upper right')
plt.show()
```

```python
plt.figure(figsize=(16, 8))
plt.xlim(0, 103)
plt.ylim(0, 1400)
plt.xlabel('age')
plt.ylabel('count')
l1, =plt.plot(male_age)
l2, =plt.plot(female_age*2.4 ,color='red')
plt.legend(handles=[l1, l2], labels=['male', 'female'],  loc='upper right')
plt.show()
```
